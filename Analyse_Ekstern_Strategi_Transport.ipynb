Nu kan vi se fordeling af tilgængelighed som funktion af gps koordinate

````
#@title fordeling af tilgængelighed som funktion af gps koordinate

import altair as alt
import pandas as pd



#Minimum_Breddegrad= alt.binding_range(min=50, max=70, step=1,name="Minimum_Breddegrad")
Laengdegrad= alt.binding_range(min=5, max=20, step=1,name="Laengdegrad")
Breddegrad= alt.binding_range(min=50, max=70, step=1,name="Breddegrad")
#Maksimum_Laengdegrad= alt.binding_range(min=5, max=20, step=1,name="Maksimum_Laengdegrad")

brush = alt.selection(type='interval', encodings=['x'], name='sel')

base=alt.Chart(data).transform_calculate(
    Laengdegrad='Laengdegrad.Laengdegrad*1',
    Breddegrad='Breddegrad.Breddegrad*1',
    #Maksimum_Laengdegrad='Maksimum_Laengdegrad.Maksimum_Laengdegrad*1',
    #Maksimum_Breddegrad='Maksimum_Breddegrad.Maksimum_Breddegrad*1'
    
).transform_filter(
  'datum.longitude>datum.Laengdegrad-0.5&&datum.longitude<datum.Laengdegrad+0.5&&datum.latitude>datum.Breddegrad-0.5&&datum.latitude<datum.Breddegrad+0.5',
  
).transform_calculate(
    Tilgaengelighed='datum.Available_charging_points/datum.Number_charging_points',
).transform_density(
    'Tilgaengelighed',
    as_=['Tilgaengelighed', 'density'],
).transform_window(
    sum='sum(density)',
    frame=[None,None]
).transform_calculate(
    sandsynlighed='datum.density/datum.sum*100',
    constant='2'
).mark_bar(color="lightgrey").encode(
    alt.X('Tilgaengelighed:Q', title = "Tilgængelighed (%)", scale=alt.Scale(domain=(0, 1))),
    alt.Y('sandsynlighed:Q', title = "Sandsynlighed (%)", scale=alt.Scale(domain=(0, 4))),

)

text = base.transform_filter(
    brush
).transform_aggregate(
    total='sum(sandsynlighed)'
).transform_calculate(
    date_range="sel.Tilgaengelighed ? sel.Tilgaengelighed[0] + ' to ' + sel.Tilgaengelighed[1] : 'all'",
    text="'Markeret område: ' + format(datum.total, '.1f')+'%'"  
).mark_text(
    align='left',
    baseline='top',
    size=16
).encode(
    x=alt.value(5),
    y=alt.value(15),
    text=alt.Text('text:N'),
)

aggregates = base.transform_aggregate(
    mean='mean(Tilgaengelighed)',
    median='median(Tilgaengelighed)',
    y_height='mean(constant)',
).transform_fold(
    ['mean', 'median']
).mark_bar().encode(
    
    x='value:Q',
    y='y_height:Q',
    #alt.Y('sandsynlighed:Q', title = "Sandsynlighed (%)", scale=alt.Scale(domain=(0, 0.1)), clamp=True),
    color='key:N'
)

(base+aggregates+text).add_selection(
    alt.selection_single(fields=['Breddegrad'], bind=Breddegrad , name='Breddegrad', init={"Breddegrad":'59'}),
    alt.selection_single(fields=['Laengdegrad'], bind=Laengdegrad , name='Laengdegrad', init={"Laengdegrad":'11'}),
    #alt.selection_single(fields=['Maksimum_Breddegrad'], bind=Maksimum_Breddegrad , name='Maksimum_Breddegrad', init={"Maksimum_Breddegrad":'70'}),
    #alt.selection_single(fields=['Maksimum_Laengdegrad'], bind=Maksimum_Laengdegrad , name='Maksimum_Laengdegrad', init={"Maksimum_Laengdegrad":'20'}),
    brush
).properties(width=600, height=300, title="Tilgængelighed af elladestander - fordeling som funktion af geografisk lokation")
````
