**Monitorering**

Vedligeholdelse af systemer forudsætter løbende monitorering af systemet. Det væsentligste formål med vedligeholdelse er løsning af specifikke udbedringsopgaver, dvs. direkte tid anvendt på vedligeholdelse. Til trods for det faktum er størrestedelen af tiden indenfor vedligeholdelse anvendt på indirekte vedligeholdelse opgaver, som handler om at forberede og koordinere vedligeholdelsesopgaver. Dette kan virke modstridende, men ikke desto mindre af pragmatiske årsager - såsom vedligeholdelsesproblemer opstår irregulært - og dertil systemer er designet med henblik mod drift og ikke vedligeholdelse, er oftest diagonisticering udfordrende.
For at sikre effektiv løsning af vedligeholdelsesopgaver kræves monitorering af systemet. Montorering består af at følge systemets tilstand over tid. Tilstanden af systemer kan oftest ikke reduceres til et enkelt øjebliks billede, og isåfald, vil sådan et billede kun give en begrænset indsigt i systemets generelle drift. For at opnå indblik i normal drift for systemet kræves monitorering af systemet. Formålet med monitoring af systemer handler grundlæggende om kontrol af systemer. Dataindsamling fra sensorer bliver oftest forvekslet med monitorering. Lagring af data er ikke tilstrækkeligt for at kunne udtale sig om et systemets tilstand. Det kræver løbende opfølgende på data og opsummering af resultater på grundlag af data. Løbende opfølgende på data er derfor en opgave i sig selv.
Overordnetset har systemer en lang række parameter, som man ønsker at påvirke i en hensigtmæssige retning. For eksempel disse parameter inkludere systemets effekt, fysiske tilstand, og teknologimæssige karakteristika mv. 
Ved at være bekendt med relevante parameter, opnår man kontrol over systemet, såledelse at man ved hvordan og i hvilken udstrækning, at man burde reagere på ændringer i en givet parameter. Sometider er systemer så komplekse at kontrol med parameter er uoverskueligt eller forståelse først skabes ved opsummering i en indikator.

Dvs. eftersom at datasættet er forbehandlet og sandsynligvis ikke normaltfordelt hjælper statistiske værktøjer blot til en forståelse, og metoderne er

````
#@title standard afvigelse af load factor
line = alt.Chart(data).mark_line(
    color='black',
    size=1
).transform_window(
    rolling_mean='mean(load_factor)',
    frame=[-20, 20]
).encode(
    x='date:T',
    y='rolling_mean:Q'
)

line_2 = alt.Chart(data).mark_errorband(opacity=1, color="black").encode(
    alt.X('date:T'),
    y=alt.Y('mean(load_factor)',scale=alt.Scale(domain=(0, 100), clamp=True)),
)

std_chart_1 = alt.Chart(data).mark_area(
    color='darkred',
    size=1,
    opacity=0.2
).transform_window(
    stdev_window='stdev(load_factor)',
    mean_window='mean(load_factor)',
    frame=[-1, 1],
).transform_calculate(
    stdev_line=alt.datum.stdev_window,
    mean_line=alt.datum.mean_window,
    oevre='datum.mean_line+datum.stdev_line',
    nedre='datum.mean_line-datum.stdev_line',
).encode(
    x='date:T',
    y='nedre:Q',
    y2='oevre:Q'
)

std_chart_2 = alt.Chart(data).mark_area(
    color='red',
    size=1,
    opacity=0.05
).transform_window(
    stdev_window='stdev(load_factor)',
    mean_window='mean(load_factor)',
    frame=[-1, 1],
).transform_calculate(
    stdev_line=alt.datum.stdev_window,
    mean_line=alt.datum.mean_window,
    oevre='datum.mean_line+2*datum.stdev_line',
    nedre='datum.mean_line-2*datum.stdev_line',
).encode(
    x='date:T',
    y='nedre:Q',
    y2='oevre:Q'
)

band = alt.Chart(data).mark_errorband(opacity=0.2, color="red",extent='ci').encode(
    alt.X('date:T'),
    y=alt.Y('load_factor'),
)

(line_2+std_chart_1+std_chart_2).properties(width=600,height=200)
````

Sometider kan programmeringssession i Google Colab fejle, og der kan være nødvendigt at genindhente datasæt.
For en sikkerhedsskyld, kan datasættet genindhentes her:

````
#@title Hent datasæt

url = 'https://raw.githubusercontent.com/jfrausig/test/master/uk_wind_farm_output_series.csv' #@param {type:"string"}

import pandas as pd
import altair as alt
data=pd.read_csv(url)
data

alt.data_transformers.disable_max_rows()
````

Ved at foretage linære regression på alle systemerne og udtrække parameteren for hældning, kan vi plotte fordeling af degradering.

````
#@title fordeling af degradering
syear = alt.binding_range(min=1990, max=2020, step=1,name="syear")


ranked_text_5 = alt.Chart(data).add_selection(
    alt.selection_single(fields=['syear'], bind=syear , name='syear', init={"syear":'2009'})
).transform_filter(
    'syear.syear*1<datum.syear'
).transform_regression(
    'yage', 'load_factor',groupby=['gen_name'], params=True
).transform_calculate(
  
  #cor_unformat='sqrt(datum.rSquared)',
  degradation_rate='datum.coef[1]',
  #cor="('Cor:'+format(datum.cor_unformat, ',.3f')+',  R2: '+format(datum.rSquared, ',.3f')+'  , b: '+format(datum.coef[1], '.3f')+'  ,a: '+format(datum.coef[0], '.1f'))",
).mark_bar(color="darkblue",opacity=0.2).encode(
    alt.X('degradation_rate:Q',title="degradation", bin=alt.Bin(extent=[-5, 5], step=0.5)),
    alt.Y('count()', stack=None,title="count"), 

)

ranked_text_5
````

Som ses foroven er en fordeling synlig. Hvis vi havde flere data, vil vi med vores domæne viden antage at fordeling kunne beskrives matematisk.
Forneden kan vi plotte den teoretiske fordeling:

````
#@title Cauchy distribution
x_0 = alt.binding_range(min=-5, max=5, step=1,name="x_0")
upsilon = alt.binding_range(min=0, max=5, step=1,name="upsilon")

data=pd.DataFrame()
data['x']=list(range(-20,20))


alt.Chart(data).mark_line().encode(
    x='x:Q',
    y='y:Q',

).transform_calculate(
    x_0='x_0.x_0*1',
    upsilon='upsilon.upsilon*1',
    y='1/(PI*datum.upsilon*(1+pow((datum.x-datum.x_0)/datum.upsilon,2))) '

).add_selection(
    alt.selection_single(fields=['upsilon'], bind=upsilon , name='upsilon', init={"upsilon":'2'}),
    alt.selection_single(fields=['x_0'], bind=x_0 , name='x_0', init={"x_0":'-1'})
)


````

Ved at sammensætte de målte degradering og den teoretiske fordeling af degradering, kan vi opsummere degraderingen på porteføljet af vindmøller og generalisere resultatet.

````
#@title Generel degradering for porteføljet
syear = alt.binding_range(min=1990, max=2020, step=1,name="syear")
x_0 = alt.binding_range(min=-5, max=5, step=1,name="x_0")
upsilon = alt.binding_range(min=0, max=5, step=1,name="upsilon")


base = alt.Chart(data).transform_filter(
    'syear.syear*1<datum.syear'
).transform_regression(
    'yage', 'load_factor',groupby=['gen_name'], params=True
).transform_calculate(
  
  cor_unformat='sqrt(datum.rSquared)',
  degradation_rate='datum.coef[1]',
  cor="('Cor:'+format(datum.cor_unformat, ',.3f')+',  R2: '+format(datum.rSquared, ',.3f')+'  , b: '+format(datum.coef[1], '.3f')+'  ,a: '+format(datum.coef[0], '.1f'))",
).transform_bin(
    'bin_degradation_rate','degradation_rate', bin=alt.Bin(extent=[-5, 5], step=0.5)
).encode(
        alt.X('bin_degradation_rate:Q', #,scale=alt.Scale(zero=False),

    )


)
degradation = base.mark_bar(color="darkblue",opacity=0.2,size=12).encode(

    alt.Y('count()', stack=None,title="count"), 

)

line=base.mark_line(color='red').encode(
    y='f_x:Q',
).transform_calculate(
    x_0='x_0.x_0*1',
    upsilon='upsilon.upsilon*1',
    x_1='datum.bin_degradation_rate',
    f_x='1/(PI*datum.upsilon*(1+pow((datum.x_1-datum.x_0)/datum.upsilon,2)))',
    #f_x='0.1'
).add_selection(
    alt.selection_single(fields=['syear'], bind=syear , name='syear', init={"syear":'2009'}),
    alt.selection_single(fields=['x_0'], bind=x_0 , name='x_0', init={"x_0":'0'}),
    alt.selection_single(fields=['upsilon'], bind=upsilon , name='upsilon', init={"upsilon":'2'})
)


alt.layer(degradation,line).resolve_scale(
    y='independent'
).properties( title="Fordeling af degraderingsrater",       padding={"left":200},width=400,height=400)
````
