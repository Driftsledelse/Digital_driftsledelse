**Monitorering**

Vedligeholdelse af systemer forudstter lbende monitorering af systemet. Det vsentligste forml med vedligeholdelse er lsning af specifikke udbedringsopgaver, dvs. direkte tid anvendt p vedligeholdelse. Til trods for det faktum er strrestedelen af tiden indenfor vedligeholdelse anvendt p indirekte vedligeholdelse opgaver, som handler om at forberede og koordinere vedligeholdelsesopgaver. Dette kan virke modstridende, men ikke desto mindre af pragmatiske rsager - ssom vedligeholdelsesproblemer opstr irregulrt - og dertil systemer er designet med henblik mod drift og ikke vedligeholdelse, er oftest diagonisticering udfordrende.
For at sikre effektiv lsning af vedligeholdelsesopgaver krves monitorering af systemet. Montorering bestr af at flge systemets tilstand over tid. Tilstanden af systemer kan oftest ikke reduceres til et enkelt jebliks billede, og isfald, vil sdan et billede kun give en begrnset indsigt i systemets generelle drift. For at opn indblik i normal drift for systemet krves monitorering af systemet. Formlet med monitoring af systemer handler grundlggende om kontrol af systemer. Dataindsamling fra sensorer bliver oftest forvekslet med monitorering. Lagring af data er ikke tilstrkkeligt for at kunne udtale sig om et systemets tilstand. Det krver lbende opflgende p data og opsummering af resultater p grundlag af data. Lbende opflgende p data er derfor en opgave i sig selv.
Overordnetset har systemer en lang rkke parameter, som man nsker at pvirke i en hensigtmssige retning. For eksempel disse parameter inkludere systemets effekt, fysiske tilstand, og teknologimssige karakteristika mv. 
Ved at vre bekendt med relevante parameter, opnr man kontrol over systemet, sledelse at man ved hvordan og i hvilken udstrkning, at man burde reagere p ndringer i en givet parameter. Sometider er systemer s komplekse at kontrol med parameter er uoverskueligt eller forstelse frst skabes ved opsummering i en indikator.

````
#@title Hent datast

url = 'https://raw.githubusercontent.com/jfrausig/test/master/uk_wind_farm_output_series.csv' #@param {type:"string"}

import pandas as pd
import altair as alt
data=pd.read_csv(url)
data

alt.data_transformers.disable_max_rows()
````

Ved at foretage linre regression p alle systemerne og udtrkke parameteren for hldning, kan vi plotte fordeling af degradering.

````
#@title fordeling af degradering
syear = alt.binding_range(min=1990, max=2020, step=1,name="syear")


ranked_text_5 = alt.Chart(data).add_selection(
    alt.selection_single(fields=['syear'], bind=syear , name='syear', init={"syear":'2009'})
).transform_filter(
    'syear.syear*1<datum.syear'
).transform_regression(
    'yage', 'load_factor',groupby=['gen_name'], params=True
).transform_calculate(
  
  #cor_unformat='sqrt(datum.rSquared)',
  degradation_rate='datum.coef[1]',
  #cor="('Cor:'+format(datum.cor_unformat, ',.3f')+',  R2: '+format(datum.rSquared, ',.3f')+'  , b: '+format(datum.coef[1], '.3f')+'  ,a: '+format(datum.coef[0], '.1f'))",
).mark_bar(color="darkblue",opacity=0.2).encode(
    alt.X('degradation_rate:Q',title="degradation", bin=alt.Bin(extent=[-5, 5], step=0.5)),
    alt.Y('count()', stack=None,title="count"), 

)

ranked_text_5
````

Som ses foroven er en fordeling synlig. Hvis vi havde flere data, vil vi med vores domne viden antage at fordeling kunne beskrives matematisk.
Forneden kan vi plotte den teoretiske fordeling:

````
#@title Cauchy distribution
x_0 = alt.binding_range(min=-5, max=5, step=1,name="x_0")
upsilon = alt.binding_range(min=0, max=5, step=1,name="upsilon")

data=pd.DataFrame()
data['x']=list(range(-20,20))


alt.Chart(data).mark_line().encode(
    x='x:Q',
    y='y:Q',

).transform_calculate(
    x_0='x_0.x_0*1',
    upsilon='upsilon.upsilon*1',
    y='1/(PI*datum.upsilon*(1+pow((datum.x-datum.x_0)/datum.upsilon,2))) '

).add_selection(
    alt.selection_single(fields=['upsilon'], bind=upsilon , name='upsilon', init={"upsilon":'2'}),
    alt.selection_single(fields=['x_0'], bind=x_0 , name='x_0', init={"x_0":'-1'})
)


````

Ved at sammenstte de mlte degradering og den teoretiske fordeling af degradering, kan vi opsummere degraderingen p portefljet af vindmller og generalisere resultatet.

````
#@title Generel degradering for portefljet
syear = alt.binding_range(min=1990, max=2020, step=1,name="syear")
x_0 = alt.binding_range(min=-5, max=5, step=1,name="x_0")
upsilon = alt.binding_range(min=0, max=5, step=1,name="upsilon")


base = alt.Chart(data).transform_filter(
    'syear.syear*1<datum.syear'
).transform_regression(
    'yage', 'load_factor',groupby=['gen_name'], params=True
).transform_calculate(
  
  cor_unformat='sqrt(datum.rSquared)',
  degradation_rate='datum.coef[1]',
  cor="('Cor:'+format(datum.cor_unformat, ',.3f')+',  R2: '+format(datum.rSquared, ',.3f')+'  , b: '+format(datum.coef[1], '.3f')+'  ,a: '+format(datum.coef[0], '.1f'))",
).transform_bin(
    'bin_degradation_rate','degradation_rate', bin=alt.Bin(extent=[-5, 5], step=0.5)
).encode(
        alt.X('bin_degradation_rate:Q', #,scale=alt.Scale(zero=False),

    )


)
degradation = base.mark_bar(color="darkblue",opacity=0.2,size=12).encode(

    alt.Y('count()', stack=None,title="count"), 

)

line=base.mark_line(color='red').encode(
    y='f_x:Q',
).transform_calculate(
    x_0='x_0.x_0*1',
    upsilon='upsilon.upsilon*1',
    x_1='datum.bin_degradation_rate',
    f_x='1/(PI*datum.upsilon*(1+pow((datum.x_1-datum.x_0)/datum.upsilon,2)))',
    #f_x='0.1'
).add_selection(
    alt.selection_single(fields=['syear'], bind=syear , name='syear', init={"syear":'2009'}),
    alt.selection_single(fields=['x_0'], bind=x_0 , name='x_0', init={"x_0":'0'}),
    alt.selection_single(fields=['upsilon'], bind=upsilon , name='upsilon', init={"upsilon":'2'})
)


alt.layer(degradation,line).resolve_scale(
    y='independent'
).properties( title="Fordeling af degraderingsrater",       padding={"left":200},width=400,height=400)
````
