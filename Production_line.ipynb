```python
textstring=[]
database=pd.DataFrame()
database.assign(**{'feature_nr':'','feature':'', 'value':''})
database_append=pd.DataFrame()
f = open("savegame.xml", "r")
for line in f:
  #print(line)
  if line[0:14]=="<data>feature_":
    database['feature']=str(line[6:].split(",")[0])
    database['value']=line[6:].split(",")[1:]
    database_append=pd.concat([database,database_append],axis=0)
    textstring.append(line)

database_append.to_csv("feature.csv")
database_append
```
